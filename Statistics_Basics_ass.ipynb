```python
# Q1. How do you calculate the mean, median, and mode of a dataset?

import statistics as stats

data = [10, 20, 30, 40, 50, 50]

mean = stats.mean(data)
median = stats.median(data)
mode = stats.mode(data)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
# Q2. Write a Python program to compute the variance and standard deviation of a dataset

import statistics as stats

data = [15, 25, 35, 45, 55]

variance = stats.variance(data)
std_dev = stats.stdev(data)

print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
# Q4. Implement sampling techniques like random sampling and stratified sampling

import pandas as pd
from sklearn.model_selection import train_test_split

# Create a sample dataset
df = pd.DataFrame({
    'Name': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Gender': ['M', 'F', 'M', 'F', 'M', 'F', 'F', 'M'],
    'Score': [50, 60, 70, 80, 90, 55, 65, 75]
})

# Simple random sampling (20%)
random_sample = df.sample(frac=0.2, random_state=1)
print("Random Sample:\n", random_sample)

# Stratified sampling based on 'Gender'
stratified_sample = df.groupby('Gender', group_keys=False).apply(lambda x: x.sample(frac=0.5, random_state=1))
print("Stratified Sample:\n", stratified_sample)
```

```python
# Q8. Generate a dataset and demonstrate positive and negative skewness

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew

# Positive skew
pos_skew = np.concatenate([np.random.normal(10, 2, 100), np.random.normal(20, 5, 20)])
# Negative skew
neg_skew = np.concatenate([np.random.normal(20, 2, 100), np.random.normal(10, 5, 20)])

# Plot
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(pos_skew, kde=True, ax=axs[0]).set_title(f"Positive Skew (Skewness = {skew(pos_skew):.2f})")
sns.histplot(neg_skew, kde=True, ax=axs[1]).set_title(f"Negative Skew (Skewness = {skew(neg_skew):.2f})")
plt.tight_layout()
plt.show()
```

```python
# Q9. Write a Python script to calculate covariance between two datasets

import numpy as np

x = [2, 4, 6, 8, 10]
y = [1, 3, 5, 7, 9]

cov_matrix = np.cov(x, y)
print("Covariance Matrix:\n", cov_matrix)
print("Covariance between x and y:", cov_matrix[0, 1])
```

```python
# Q10. Write a Python script to calculate the correlation coefficient between two datasets

import numpy as np

x = [2, 4, 6, 8, 10]
y = [1, 3, 5, 7, 9]

correlation = np.corrcoef(x, y)
print("Correlation Matrix:\n", correlation)
print("Correlation Coefficient:", correlation[0, 1])
```

```python
# Q12. Implement and compare simple random sampling and systematic sampling

import numpy as np
import pandas as pd

data = pd.DataFrame({'Value': range(1, 21)})

# Simple Random Sampling
random_sample = data.sample(n=5, random_state=42)

# Systematic Sampling (every 4th element)
systematic_sample = data.iloc[::4]

print("Simple Random Sample:\n", random_sample)
print("Systematic Sample:\n", systematic_sample)
```

```python
# Q13. Calculate the mean, median, and mode of grouped data

import pandas as pd
from statistics import mean, median, mode

# Assume frequency table (grouped data)
grouped_data = pd.DataFrame({
    'Class Interval': ['0-10', '10-20', '20-30'],
    'Frequency': [5, 10, 5]
})

# Approximate using midpoints
midpoints = [5, 15, 25]
frequencies = [5, 10, 5]
total = sum(frequencies)

group_mean = sum([m * f for m, f in zip(midpoints, frequencies)]) / total
print("Approximate Mean (Grouped):", group_mean)
```

```python
# Q24. Implement the Pearson and Spearman correlation coefficients for a dataset

from scipy.stats import pearsonr, spearmanr

x = [10, 20, 30, 40, 50]
y = [15, 25, 35, 45, 60]

pearson_corr, _ = pearsonr(x, y)
spearman_corr, _ = spearmanr(x, y)

print("Pearson Correlation:", pearson_corr)
print("Spearman Correlation:", spearman_corr)
```
