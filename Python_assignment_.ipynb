```python
# 1. Function Example
def greet():
    print("Hello!")
greet()

# Method Example
name = "Alice"
print(name.upper())  # upper() is a method

# 2. Function Arguments and Parameters
def add(a, b):  # a and b are parameters
    return a + b
print(add(2, 3))  # 2 and 3 are arguments

# 3. Ways to Define and Call a Function
def square(x):
    return x * x
print(square(4))

cube = lambda x: x**3
print(cube(2))

# 4. Purpose of return
def multiply(x, y):
    return x * y
result = multiply(3, 5)
print(result)

# 5. Iterators vs Iterables
lst = [1, 2, 3]
it = iter(lst)  # iterator
print(next(it))
print(next(it))

# 6. Generators
def gen_numbers():
    for i in range(3):
        yield i
for num in gen_numbers():
    print(num)

# 7. Advantages of Generators
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1
for num in count_up_to(5):
    print(num)

# 8. Lambda Function
add = lambda a, b: a + b
print(add(3, 5))

# 9. map()
nums = [1, 2, 3]
squared = list(map(lambda x: x**2, nums))
print(squared)

# 10. sum of even numbers
def sum_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)
print(sum_even_numbers([1, 2, 3, 4, 5, 6]))

# 11. Reverse a string
def reverse_string(s):
    return s[::-1]
print(reverse_string("hello"))

# 12. Squares of list elements
def square_list(numbers):
    return [num ** 2 for num in numbers]
print(square_list([1, 2, 3]))

# 13. Check prime numbers from 1 to 200
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
primes = [n for n in range(1, 201) if is_prime(n)]
print(primes)

# 14. Fibonacci Iterator Class
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a, self.b = 0, 1
    def __iter__(self):
        return self
    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        self.count += 1
        value = self.a
        self.a, self.b = self.b, self.a + self.b
        return value
for num in FibonacciIterator(7):
    print(num, end=' ')

# 15. Generator for powers of 2
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i
for power in powers_of_two(5):
    print(power, end=' ')

# 16. Generator to read file line by line
def read_file_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()
# Example usage commented
# for line in read_file_lines("sample.txt"):
#     print(line)

# 17. Sort tuples by second element using lambda
tuples = [(1, 3), (2, 1), (4, 2)]
sorted_tuples = sorted(tuples, key=lambda x: x[1])
print(sorted_tuples)

# 18. Convert Celsius to Fahrenheit using map()
celsius = [0, 20, 37, 100]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)

# 19. Remove vowels from string using filter()
def remove_vowels(s):
    return ''.join(filter(lambda c: c.lower() not in 'aeiou', s))
print(remove_vowels("Hello World"))

# 20. Book Shop Order Accounting with Lambda and map()
orders = [
    ["34587", "Learning Python", 40.95, 1],
    ["98762", "Programming Python", 56.80, 1],
    ["77226", "Head First Python", 32.95, 3],
    ["88112", "EinfÃ¼hrung in Python3", 24.99, 1]
]
order_totals = list(map(
    lambda order: (order[0], order[2]*order[3] if order[2]*order[3] >= 100 else order[2]*order[3] + 10),
    orders
))
print(order_totals)
```
